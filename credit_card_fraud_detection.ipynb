Dealing with unbalanced data

Building a sample dataset containing similar distribution of legit and fraudulent transactions
No. of legit transactions--> 553574
No. of fraudulent transactions--> 2145

joining it to make normal distribution

```python
#Random Sampling
legit_sample_data= legit.sample(n=2145)
```

```python
#Concatenating the dataframes for Normal Distribution
#Adding ROW WISE AXIS=0
#COLUMN WISE AXIS=1
concat_dataframes= pd.concat([legit_sample_data], axis=0)
c_d= concat_dataframes
```

```python
#Spliting the into features(independent variables) and targets(dependent variable)
X= c_d.drop(columns= 'is_fraud', axis=1)
Y= c_d['is_fraud']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Spliting the data into Training and Testing data
X_train, X_test, Y_train, Y_test = tts(X, Y, test_size=0.2,stratify= Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
Y_train = Y_train.astype(float)
```

```python
X_train['trans_date_trans_time'] = pd.to_datetime(X_train['trans_date_trans_time'])

# Extract relevant features from the timestamp column
X_train['year'] = X_train['trans_date_trans_time'].dt.year
X_train['month'] = X_train['trans_date_trans_time'].dt.month
X_train['day'] = X_train['trans_date_trans_time'].dt.day
X_train['hour'] = X_train['trans_date_trans_time'].dt.hour
X_train['minute'] = X_train['trans_date_trans_time'].dt.minute

# Drop the original timestamp column if not needed for modeling
X_train.drop('trans_date_trans_time', axis=1, inplace=True)
```

```python
#Training the model usning Logistic Regression for binary classification
model_lr= LogisticRegression()
```

```python
# Fit the model using the preprocessed features (X_train) and Y_train
model_lr.fit(X_train.values, Y_train.values)
```
